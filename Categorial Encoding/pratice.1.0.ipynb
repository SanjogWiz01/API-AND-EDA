```python
import pandas as pd
import category_encoders as ce

data = {"City":["Pokhara","Kathmandu","Butwal","Pokhara"]}

df = pd.DataFrame(data)

encoder = ce.BinaryEncoder(cols=["City"])

encoded = encoder.fit_transform(df)

print(encoded)
```

```python
freq = df["City"].value_counts()

df["City_freq"] = df["City"].map(freq)
```

```python
mean_encoding = df.groupby("City")["Bought"].mean()

df["City_mean"] = df["City"].map(mean_encoding)
```

```python
encoder = ce.HashingEncoder(cols=["City"], n_components=4)
df_hash = encoder.fit_transform(df)
```
